```python
#chose your variables (we are using plt)
height = best_y09["User Rating"]
bars = best_y09["Name"]
y_pos = np.arange(len(bars))

# Create bars
plt.bar(y_pos, height)

# Create names on the x-axis
plt.xticks(y_pos, bars)

#resize
figsize=(10, 40), dpi=80

# Show graphic
plt.show()
```

```python
#try horizontal bc yes
height = best_y09["User Rating"]
bars = best_y09["Name"]
y_pos = np.arange(len(bars))
 
# Create horizontal bars
plt.barh(y_pos, height)
 
# Create names on the x-axis
plt.yticks(y_pos, bars)
 
# Show graphic
plt.show()
```
